```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()

x_train_std = scalar.fit_transform(x_train)
x_test_std = scalar.fit_transform(x_test)
```

```python
from sklearn.decomposition import PCA
pca =PCA()
pca.fit(x_train_std)
```

```python
var =  pca.explained_variance_ratio_
var
```

```python
np.sum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
```

```python
var1 = np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
```

```python
var1
```

```python
pca.explained_variance_ratio_[:22].sum()
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumlative explained variance')
plt.annotate('22',xy=(22,95))
```

```python
pca_model = PCA(n_components=22)
```

```python
train_pca = pca_model.fit_transform(x_train_std)
test_pca = pca_model.fit_transform(x_test_std)
```

```python
plt.figure(figsize=(10,10))
correlation = pd.DataFrame(train_pca).corr()
sns.heatmap(correlation,vmax=1,square =True,cmap='summer')
plt.title('correlation btw different features')
```
